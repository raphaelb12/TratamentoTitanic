```python
import pandas as pd
```

```python
dados = pd.read_csv('titanic.csv') #recebe o arquivo
```

```python
dados = dados.drop(columns = ['SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked', 'PassengerId']) #drop de colunas nao desejadas
dados
```

```python
dados.columns #retorna as colunas restantes
```

```python
dados2 = dados[['Name' ,'Age', 'Sex', 'Pclass', 'Survived']] #organiza as colunas na ordem desejada
dados2
```

```python
filtros = dados2.loc[(dados2['Sex'] == 'female') & (dados2['Pclass'] == 1) & (dados2['Survived'] == 1) ] #faz o filtro na ordem sexo feminino -> se estava na primeira classe -> se sobreviveu
filtros
```

```python
filtros = filtros.sort_values('Name') #coloca na ordem alfabetica
filtros
```

```python
filtros.to_csv('ProjetoTitanic2.csv', index = False) #transforma em um arquivo csv
```
